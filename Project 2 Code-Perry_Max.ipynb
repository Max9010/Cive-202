```python
# CIVE 202 - Project 2
# Group G
# Concrete Mix Design Calculator

import numpy as np
import pandas as pd


# --- functions ---

def get_water(cem, fly, sil, other, ratio):
    total = cem + fly + sil + other
    water = total * ratio
    return water


def get_volume(weight, sg):
    return weight / (sg * 62.4)


def air_vol(percent):
    return (percent / 100) * 27


def total_agg_vol(vc, vf, vs, vo, va, vw):
    return 27 - vc - vf - vs - vo - va - vw


def agg_weight(percent, sg, volume):
    return 62.4 * (percent / 100) * sg * volume


# --- user input section ---

project_number = int(input("Project Number: "))
concrete_type = input("Concrete Class: ")
control_number = int(input("Control Number: "))

cement = float(input("Cement (lb/yd3): "))
flyash = float(input("Fly Ash (lb/yd3): "))
silica = float(input("Silica Fume (lb/yd3): "))
other_scm = float(input("Other SCM (lb/yd3): "))

wc_ratio = float(input("Water/Cement Ratio: "))
air_percent = float(input("Air Content (%): "))

fine_pct = float(input("Fine Aggregate %: "))
coarse_pct = float(input("Coarse Aggregate %: "))
other_pct = float(input("Other Aggregate %: "))

cement_sg = float(input("Cement SG: "))
flyash_sg = float(input("Fly Ash SG: "))
silica_sg = float(input("Silica SG: "))
other_sg = float(input("Other SCM SG: "))
fine_sg = float(input("Fine Aggregate SG: "))
coarse_sg = float(input("Coarse Aggregate SG: "))
other_agg_sg = float(input("Other Aggregate SG: "))

print("\nInputs saved for Group G\n")


# --- calculations ---

water_needed = get_water(cement, flyash, silica, other_scm, wc_ratio)

vol_cement = get_volume(cement, cement_sg)
vol_flyash = get_volume(flyash, flyash_sg)
vol_silica = get_volume(silica, silica_sg)
vol_other = get_volume(other_scm, other_sg)

vol_air = air_vol(air_percent)
vol_water = water_needed / 62.4

aggregate_volume = total_agg_vol(
    vol_cement,
    vol_flyash,
    vol_silica,
    vol_other,
    vol_air,
    vol_water
)

fine_weight = agg_weight(fine_pct, fine_sg, aggregate_volume)
coarse_weight = agg_weight(coarse_pct, coarse_sg, aggregate_volume)
other_weight = agg_weight(other_pct, other_agg_sg, aggregate_volume)


# --- output ---

print("===================================")
print("Concrete Mix Results - Group G")
print("===================================")

print("Project:", project_number)
print("Class:", concrete_type)
print("Control:", control_number)
print("-----------------------------------")

print("Cement:", round(cement,2), "lb")
print("Fly Ash:", round(flyash,2), "lb")
print("Silica Fume:", round(silica,2), "lb")
print("Other SCM:", round(other_scm,2), "lb")

print("Fine Aggregate:", round(fine_weight,2), "lb")
print("Coarse Aggregate:", round(coarse_weight,2), "lb")
print("Other Aggregate:", round(other_weight,2), "lb")

print("Water:", round(water_needed,2), "lb")

print("-----------------------------------")
print("Done - Group G")
```
